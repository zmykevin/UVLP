```python
import json
import os

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load the Clip_Score Json
data_path = "/data/home/zmykevin/vinvl_data/CC"
with open(os.path.join(data_path, "cc_clip_score_tfidf_sorted.json"), "r") as f:
    cc_clip_score = json.load(f)
```

```python
clip_scores = []
for k,x in cc_clip_score.items():
    clip_scores.append(x)
clip_scores = np.array(clip_scores)
```

```python
#Plot the histogram
n,bins,patches = plt.hist(clip_scores, 20)
plt.xlabel('clip_score')
plt.ylabel('sample_numbers')
plt.title('Histogram of clip_scores for {}'.format("tfidf retrived sorted captions"))
plt.show()
```
